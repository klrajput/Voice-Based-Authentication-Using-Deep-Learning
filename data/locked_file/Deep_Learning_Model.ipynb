```python
# ANN
from tensorflow import keras
mnist = keras.datasets.mnist # load the mnist dataset from keras
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[173], cmap='grey')
```

```python
# define the neural network
model = keras.models.Sequential()

# add artificial neurons in layers
model.add(keras.layers.Flatten()) # gives 28*28 = 784 neurons
# hidden layer
model.add(keras.layers.Dense(540, input_shape=(28, 28), activation='relu'))
# output layer
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
# loss OPTIMISER
```

```python
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=15)
```
